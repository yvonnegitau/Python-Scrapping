```
def scapdata(year):
    page = requests.get("http://www.resultdb.com/english-premier-league-tables/"+year+"/")
    soup = BeautifulSoup(page.text,'lxml')
    table = soup.find('table')

    data = []
    rows = table.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele]) # Get rid of empty values

    columns= ['position','team name','games','won','draw','lost','goal scored','goals conceded','goal difference','points']
    season = pd.DataFrame(data[1:],columns=columns)
    season['avg_goals_per_game'] = pd.to_numeric(season['goal scored'])/pd.to_numeric(season['games'])
    season['avg_goals_conceded_per_game'] = pd.to_numeric(season['goals conceded'])/pd.to_numeric(season['games'])
    season['season'] = year
    return season
```

```
years = ['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009',
         '2010','2011','2012','2013','2014','2015']
history = []
for year in years:
    season = scapdata(year)
    history.append(season)
seasonal_data = pd.concat(history)
```
